# Sorting pandas DataFrames or Series

```python
dataframe_csv = pd.read_csv('little_rock.csv')
dataframe_csv.head()
```

## Series Sorting

## assecending Order by default

```python
dataframe_csv['vehicle_type'].sort_values()
# bracket notation selection of series
```

## Dessecending Order

```python
dataframe_csv.vehicle_type.sort_values(ascending=False)
# dot notation select of series
```

## dataframe sorting

## Ascending Order

```python
dataframe_csv.sort_values('vehicle_type').head()
```

## Descending Order

```python
dataframe_csv.sort_values('subject_age', ascending=False).head()
```

## Sorting more than one columns

```python
dataframe_csv.sort_values(['subject_age', 'vehicle_type'])
```

# Appling multiple Filtering like (Select clause in Sql)

## Using single statement for filtering statement

```python
dataframe_csv[dataframe_csv.subject_age <= 16.0]
```

## AND & OR

## Filtering dataframe with AND statement

```python
dataframe_csv[(dataframe_csv.subject_age <= 16.0) & (dataframe_csv.subject_sex == 'male')]
```

##  Using OR statement (OR)

```python
dataframe_csv[(dataframe_csv.raw_row_number == '18097') | (dataframe_csv.subject_age <= 16.0)]
```

## Selecting single column of dataframe

```python
dataframe_csv.date[(dataframe_csv.date == '2017-01-07') & (dataframe_csv.subject_race == 'white')]
```

## Filtering one column with different conditions

```python
dataframe_csv[(dataframe_csv.subject_age == 14.0) | (dataframe_csv.subject_age == 15.0) | (dataframe_csv.subject_age == 16.0 )]
```

### Another Method

```python
dataframe_csv[dataframe_csv.subject_age.isin([14.0,15.0,16.0])]
```
