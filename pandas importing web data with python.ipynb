```python
#this is how we import our html data from web site through urllib package

import pandas as pd
from urllib.request import urlopen, Request
url = 'https://www.datacamp.com/'
request = Request(url)
response = urlopen(request)
html = response.read()
print(type(html))
response.close()
```

```python
#In this section we will import html data from web through Requests packages

import requests

url = 'https://www.datacamp.com/'
r = requests.get(url)
text = r.text
print(text)
```

```python
#this is how we import html data from web through beautifull soup
#beautisful soup use for to show data with tags and clear

from bs4 import BeautifulSoup

url = 'https://www.python.org/~guido/'
r = requests.get(url)
html = r.text
soup = BeautifulSoup(html)
pretty_soup = soup.prettify()

#this is how we retrieve sepicfic tags or title or others html elementss data from html web page 

group_title = soup.title
print(group_title)
group_text = soup.get_text()
# print(group_text)


a_tags = soup.find_all('a')

for link in a_tags:
    print(link.get('href'))
```
